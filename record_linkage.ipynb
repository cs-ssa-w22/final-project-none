```python
import pandas as pd
import re
import math

apt_filename = "apartments.csv"
domu_filename = "domu.csv"
apt = pd.read_csv(apt_filename, dtype='string')
domu = pd.read_csv(domu_filename, dtype='string')
```

Change Apartments and Domu Dataframe into consistent formatting

```python
apt = apt.replace({"room_area": ","}, {"room_area": ""}, regex=True)
```

```python
apt
```

```python
domu["bedroom"] = domu["bedroom"].str.lower()
domu = domu.replace({"bedroom": "0 - 0 bed"}, {"bedroom": "Studio"}, regex=True)
domu = domu.replace({"bedroom": "0"}, {"bedroom": "Studio"}, regex=True)
domu = domu.replace({"bedroom": "bed"}, {"bedroom": "bd"}, regex=True)
domu = domu.replace({"address": "Chicago, IL,"}, {"address": "Chicago IL"}, regex=True)
domu = domu.replace({"amenities": "In-unit Laundry"}, {"amenities": "In Unit Washer & Dryer"}, regex=True)

domu["bathroom"] = domu["bathroom"].str.lower()
domu = domu.replace({"bathroom": "bath"}, {"bathroom": "ba"}, regex=True)
```

```python
domu
```

Combine the two dataframes

```python
housing = pd.DataFrame(columns=[
    "property_name", 
    "address", 
    "latitude", 
    "longitude", 
    "price_range", 
    "bedroom", 
    "bathroom", 
    "room_area",
    "amenities"]
    )
```

```python
domu_lst = []

for a, arow in apt.iterrows():
    for d, drow in domu.iterrows():
        address_apt = (arow[1].replace(',', '')).lower()
        address_domu = (drow[1].replace(',', '')).lower()
        #bacause only domu has patters like '#' or '-'
        d_add = re.findall(r'#\w+|-\d+', drow[1]) 
        if d_add != []:
            for patterns in d_add:
                address_domu = address_domu.replace(d_add[0].lower(), '')
            address_domu = re.sub(r'[^\w]+', ' ',address_domu).strip() 

        if address_apt == address_domu:
            domu_lst.append(d)

            #check price_range
            if arow["price_range"] == "Call for Rent":
                aprice_low, aprice_high = math.inf, -math.inf
            else:
                aprice = [float(v) for v in arow["price_range"].replace("$", "").replace(",", "").split(" - ")]
                aprice_low, aprice_high = aprice[0], aprice[-1]
            dprice = [float(v) for v in drow["price_range"].replace("$", "").replace(",", "").split("–")]
            dprice_low, dprice_high = dprice[0], dprice[-1]

            if dprice_low < aprice_low:
                price_low = drow["price_range"].split("–")[0]
            else:
                price_low = arow["price_range"].split(" - ")[0]
            
            if dprice_high > aprice_high:
                price_high = drow["price_range"].split("–")[-1]
            else:
                price_high = arow["price_range"].split(" - ")[-1]
            
            if price_low != price_high:
                arow["price_range"] = str(price_low) + " - " + str(price_high)
            else:
                arow["price_range"] = str(price_low)

            #check bedroom
            abd = [int(v) for v in arow["bedroom"].replace("Studio", "0").replace(" bd", "").split(" - ")]
            dbd = [int(v) for v in drow["bedroom"].replace("Studio", "0").replace(" + den", "").replace(" bd", "").split(" - ")]
            abd_low, abd_high = abd[0], abd[-1]
            dbd_low, dbd_high = dbd[0], dbd[-1]

            if min(abd_low, dbd_low) == max(abd_high, dbd_high):
                arow["bedroom"] = str(min(abd_low, dbd_low))
            else:
                arow["bedroom"] = str(min(abd_low, dbd_low)) + " - " + str(max(abd_high, dbd_high)) + " bd"
            if " + den" in drow["bedroom"]:
                arow["bedroom"] += " + den"
            arow ["bedroom"] = arow["bedroom"].replace("0 bd", "Studio").replace("0", "Studio")

            #check bathroom
            aba = [float(v) for v in arow["bathroom"].replace(" ba", "").split(" - ")]
            dba = [float(v) for v in drow["bathroom"].replace(" ba", "").split(" - ")]
            aba_low, aba_high = aba[0], aba[-1]
            dba_low, dba_high = dba[0], dba[-1]

            if min(aba_low, dba_low) == max(aba_high, dba_high):
                arow["bathroom"] = str(min(aba_low, dba_low))
            else:
                arow["bathroom"] = str(min(aba_low, dba_low)) + " - " + str(max(aba_high, dba_high)) + " ba"

            #check room_area
            if not pd.isnull(arow["room_area"]):
                aarea = [int(v) for v in arow["room_area"].replace(" sq ft", "").split(" - ")]
                aarea_low, aarea_high = aarea[0], aarea[-1]
            else:
                aarea_low, aarea_high = math.inf, -math.inf
            darea = [int(v) for v in drow["room_area"].replace(" sq ft", "").split(" - ")]
            darea_low, darea_high = darea[0], darea[-1]

            if min(aarea_low, darea_low) == max(aarea_high, darea_high):
                arow["room_area"] = str(min(aarea_low, darea_low)) + " sq ft"
            else:
                arow["room_area"] = str(min(aarea_low, darea_low)) + " - " + str(max(aarea_high, darea_high)) + " sq ft"

            #check amenities
            apt_amenities = arow["amenities"].split(", ")
            domu_amenities = drow["amenities"].split(", ")
            amenities = list(set(apt_amenities + domu_amenities))
            arow["amenities"] = ", ".join(amenities)
    
    housing.loc[housing.shape[0]] = arow
        
```

```python
housing = pd.concat([housing, domu.drop(domu_lst)])
housing.to_csv("combined_housing.csv", index=False)
```
