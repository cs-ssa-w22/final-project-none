```python
import csv
import jellyfish
import pandas as pd
import re


apt_filename = "apartments.csv"
domu_filename = "domu.csv"
apt = pd.read_csv(apt_filename, dtype='string')
domu = pd.read_csv(domu_filename, dtype='string')
```

```python
lst = []
known_list = []
for a, arow in apt.iterrows():
    for d, drow in domu.iterrows():
        address_apt = (arow[1].replace(',', '')).lower()
        address_domu = (drow[1].replace(',', '')).lower()
        #bacause only domu has patters like '#' or '-'
        d_add = re.findall(r'#\w+|-\d+', drow[1]) 
        if d_add != []:
            for patterns in d_add:
                address_domu = address_domu.replace(d_add[0].lower(), '')
            address_domu = re.sub(r'[^\w]+', ' ',address_domu).strip()            
        if address_apt == address_domu:
            lst.append([arow[1], drow[1]])
```
